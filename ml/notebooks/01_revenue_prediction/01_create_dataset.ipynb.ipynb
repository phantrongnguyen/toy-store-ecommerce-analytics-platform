```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fact = pd.read_csv("data/processed/fact.csv")
dim_session = pd.read_csv("data/processed/dim_session.csv")
dim_product = pd.read_csv("data/processed/dim_product.csv")
dim_order = pd.read_csv("data/processed/dim_order.csv")
dim_time = pd.read_csv("data/processed/dim_time.csv")
```

## 1. Join tables

### 1.1. Join dim_session

```python
df = fact.merge(
    dim_session,
    on="website_session_id",
    how="left"
)
```

### 1.2. Join dim_session 

```python
df = df.merge(
    dim_product,
    on="product_id",
    how="left"
)
```

```python
df.head(10)
```

```python
df.shape
```

### 1.3. Join dim_order

```python
df = df.merge(
    dim_order,
    on="order_id",
    how="left"
)
```

```python
df.shape
```

### 1.4. Join dim_time

```python
df = df.merge(
    dim_time,
    on="time_id",
    how="left"
)
```
