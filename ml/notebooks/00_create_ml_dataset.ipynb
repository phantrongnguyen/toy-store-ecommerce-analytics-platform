## 1. **Import libary**

```python
import pyodbc
import pandas as pd
from sqlalchemy import create_engine
import urllib
```

```python
pyodbc.drivers()
```

```python
SERVER = 'LENOVO\SQLEXPRESS'
DATABASE = 'test'
DRIVER = 'ODBC Driver 17 for SQL Server'
```

```python
# 1. Cấu hình chuỗi kết nối
# Chúng ta cần quote lại chuỗi kết nối để xử lý các ký tự đặc biệt như dấu gạch chéo ngược (\)
params = urllib.parse.quote_plus(
    f'DRIVER={{{DRIVER}}};'
    f'SERVER={SERVER};'
    f'DATABASE={DATABASE};'
    f'Trusted_Connection=yes;'
)

# 2. Khởi tạo Engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```
