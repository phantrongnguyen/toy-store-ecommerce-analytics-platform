## 1. **Import libary**

```python
import pyodbc
import pandas as pd
from sqlalchemy import create_engine
import urllib
```

```python
pyodbc.drivers()
```

```python
SERVER = 'LENOVO\SQLEXPRESS'
DATABASE = 'toy_store_datawarehouse'
DRIVER = 'ODBC Driver 17 for SQL Server'
```

```python
# 1. Cấu hình chuỗi kết nối
# Chúng ta cần quote lại chuỗi kết nối để xử lý các ký tự đặc biệt như dấu gạch chéo ngược (\)
params = urllib.parse.quote_plus(
    f'DRIVER={{{DRIVER}}};'
    f'SERVER={SERVER};'
    f'DATABASE={DATABASE};'
    f'Trusted_Connection=yes;'
)

# 2. Khởi tạo Engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# 3. Viết câu lệnh SQL lấy dữ liệu
# Thay 'YourTableName' bằng tên bảng thực tế trong toy_store_datawarehouse
query = "SELECT TOP 100 * FROM stage" 

try:
    # 4. Đọc dữ liệu vào DataFrame
    df = pd.read_sql(query, engine)
    
    # Hiển thị kết quả
    print("Kết nối thành công! Dưới đây là dữ liệu:")
    print(df.head())

except Exception as e:
    print(f"Có lỗi xảy ra: {e}")
```
