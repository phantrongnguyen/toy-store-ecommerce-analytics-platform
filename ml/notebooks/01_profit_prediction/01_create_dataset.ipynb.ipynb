```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fact = pd.read_csv("data/processed/fact.csv")
dim_session = pd.read_csv("data/processed/dim_session.csv")
dim_product = pd.read_csv("data/processed/dim_product.csv")
dim_order = pd.read_csv("data/processed/dim_order.csv")
dim_time = pd.read_csv("data/processed/dim_time.csv")
```

## 1. Join tables

### 1.1. Join dim_session

```python
df = fact.merge(
    dim_session,
    on="website_session_id",
    how="left"
)
```

### 1.2. Join dim_session 

```python
df = df.merge(
    dim_product,
    on="product_id",
    how="left"
)
```

### 1.3. Join dim_order

```python
df = df.merge(
    dim_order,
    on="order_id",
    how="left"
)
```

### 1.4. Join dim_time

```python
df = df.merge(
    dim_time,
    on="time_id",
    how="left"
)
```

## 3. Feature Engineering

### 3.1. Create profit column

```python
df['profit'] = df['price_usd'] - df['cogs_usd']
```

```python
df['profit'].describe()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["profit"], bins=30)
plt.title("Distribution of Profit")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

```python
# choose features and target variable

target = "profit"

features = [
    "items_purchased",
    "product_id",
    "product_name",
    "primary_product_id",
    "is_primary_item",
    "utm_source",
    "utm_campaign",
    "utm_content",
    "device_type",
    "http_referer",
    "is_repeat_session",
    "year",
    "month",
    "day",
    "quater",
    "hour"
]

ml_df = df[features + [target]].copy()

ml_df.head()
```

```python
# handle missing values

categorical_cols = [
    "product_name",
    "utm_source",
    "utm_campaign",
    "utm_content",
    "device_type",
    "http_referer"
]

numerical_cols = [
    "items_purchased",
    "product_id",
    "primary_product_id",
    "is_primary_item",
    "is_repeat_session",
    "year",
    "month",
    "day",
    "quater",
    "hour"
]

for col in categorical_cols:
    ml_df[col] = ml_df[col].fillna("unknown")

for col in numerical_cols:
    ml_df[col] = ml_df[col].fillna(0)

ml_df[target] = ml_df[target].fillna(0)
ml_df = ml_df[ml_df[target] >= 0]

ml_df.isnull().sum()
```

```python
# save processed dataset

ml_df.to_csv("data/interim/processed_profit_data.csv", index=False)

ml_df.shape
```
